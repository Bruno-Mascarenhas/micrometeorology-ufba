```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from scipy import stats
plt.style.use('seaborn-whitegrid')
from sys import exit
import scipy
```

```python
labmim = pd.read_csv('labmim_completo_controle2020.dat',sep=';')
labmim.index = pd.to_datetime(labmim[['year','month','day','hour']])

labmim2 = pd.read_csv('labmim_2020_2021.dat',sep=';')
labmim2.index = pd.to_datetime(labmim2[['year','month','day','hour']])
```

```python
mapper = {'CM3Up_Wm2_Avg':'Sw_dw',
        'CM3Dn_Wm2_Avg':'Sw_up',
        'CG3Up_Wm2Cr_Avg':'Lw_dw',
        'CG3Dn_Wm2Cr_Avg':'Lw_up',
        'CMP21_Wm2_Avg':'Sw_dif'}

labmim2.rename(columns=mapper,inplace=True)
```

```python
labmim2 = labmim2['year month day hour Sw_dif Sw_dw Sw_up Lw_dw Lw_up'.split()]
labmim = labmim['year month day hour Sw_dif Sw_dw Sw_up Lw_dw Lw_up'.split()]
```

```python
final = pd.merge_ordered(labmim, labmim2)
final.index = pd.to_datetime(final[['year','month','day','hour']])
```

```python
final.to_csv('balanco_media_horaria.dat',na_rep='nan',sep=';',index=False)
```

```python
tmp = final.groupby(['year','month'])
```

```python
x = tmp.get_group((2021,1))['Sw_dw']
x[x>0].mean()
```

```python
balanco = []
```

```python
for year in range(2016,2022):
    for month in range(1,13):
        try:
            swdif = tmp.get_group((year,month))['Sw_dif']
            swdif = swdif[swdif>0].mean()
            
            swdw = tmp.get_group((year,month))['Sw_dw']
            swdw = swdw[swdw>0].mean()
            
            swup = tmp.get_group((year,month))['Sw_up']
            swup = swup[swup>0].mean()
            
            lwup = tmp.get_group((year,month))['Lw_dw']
            lwup = lwup[lwup>0].mean()
            
            lwdw = tmp.get_group((year,month))['Lw_up']
            lwdw = lwdw[lwdw>0].mean()
            
            balanco.append([year,month,swdif,swdw,swup,lwdw,lwup])
        except Exception as err:
            #print(err)
            continue
```

```python
balanco = pd.DataFrame(data=balanco,columns=['year','month','Sw_dif','Sw_dw','Sw_up','Lw_dw','Lw_up'])
```
